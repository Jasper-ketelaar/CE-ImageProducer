```python
def getImageNames(img):
    dimensions = ['64','176','200','512']
    img_format = ['jpg','png']
    output = []
    for dindex in range(len(dimensions)):
        temp = img.split('_')
        temp[2] = '7'  # set rotation angle here 0-11
        temp[3] = dimensions[dindex]
        temp = '_'.join(temp)
        for findex in range(len(img_format)):
            temp = temp.split('.')
            temp[1] = img_format[findex]
            temp = '.'.join(temp)
            output.append(temp)
    return output

getImageNames('a_0_0_512_0_0.png')
```

```python
count = 0 
for index in range(len(data)):
    img_link = data[index]['image_front'].split('/')
    image_names = getImageNames(img_link.pop())
    image_found = False
    for iindex in range(len(image_names)):
        url = '/'.join(img_link) + '/' + image_names[iindex]
        response = requests.get(url)
        if(response.status_code == 200):
            image_found = True
            count = count + 1
            file = open('./wine_images/'+ data[index]['sku'] + '/' + image_names[iindex], "wb")
            file.write(response.content)
            file.close()
            break
    if (image_found == False):
        print(img_link)
print(count)
```

```python
import os
# For blurred images
base = './wine_images/'
count = 0
for index in range(len(sku)):
    dir_name = os.path.join(base, sku[index])
    for file in os.listdir(dir_name):
        img_name = file.split('_')
        if(img_name[1] != '1'):
            OriImage = Image.open(os.path.join(dir_name,file))
            #OriImage.show()

            boxImage = OriImage.filter(ImageFilter.BLUR)
            #boxImage.show()


            img_name[1] = '1'
            count = count+1
            boxImage.save(os.path.join(dir_name, '_'.join(img_name)))

print(count)
```

```python
#resize with width 448
images = [file for file in os.listdir() if file.endswith(('jpeg', 'png', 'jpg'))]
for image in images:
    img = Image.open(image)
    img.thumbnail((448,448))
    img.save("resized_"+image, optimize=True, quality=40)
```

```python
#resize with height and width 448
image = Image.open('a_0_7_176_0_0.png')
print(image.size)

resized_image = image.resize((448,448))
resized_image.save('448.png')
print(resized_image.size)
```

```python
# resize with height 448
fixed_height = 448
image = Image.open('a_0_7_176_0_0.png')
height_percent = (fixed_height / float(image.size[1]))
width_size = int((float(image.size[0]) * float(height_percent)))
image = image.resize((width_size, fixed_height), Image.NEAREST)
image.save('resized_nearest.png')
```

```python
import cv2
import numpy as np
import os
 
img = cv2.imread('a_0_7_176_0_0.png')

gauss = np.random.normal(0,1,img.size)
gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
noise = img + img * gauss

cv2.imwrite(os.path.join('./' , 'waka.jpg'), noise)
cv2.imshow('a',noise)
cv2.waitKey(0)
```

```python
import os
# For noisy images
base = './wine_images/'
count = 0
for index in range(len(sku)):
    dir_name = os.path.join(base, sku[index])
    for file in os.listdir(dir_name):
        img_name = file.split('.')[0] + '_n'
        img = cv2.imread(os.path.join(dir_name,file))
        gauss = np.random.normal(0,1,img.size)
        gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
        noise = img + img * gauss

        cv2.imwrite(os.path.join(dir_name , img_name+'.'+ file.split('.')[1]), noise)
        count = count+1

print(count)
```
